#cropping

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
image = cv2.imread("/content/drive/MyDrive/images/input.jpg")
plt.imshow(image[...,::-1])
```

```python
height, width = image.shape[:2]

# Simply use indexing to crop out the rectangle we desire
cropped = image[15:75 , 230:300]

plt.imshow(cropped[...,::-1])
```

```python
cropped_wh = image[height // 2 : 2 * height // 2, width // 2 : 2 * width // 2]
```

```python
plt.imshow(cropped_wh[...,::-1])
```
